```python
#separate dependent and independent variable
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#splitting into training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

# Apply various model to predict the accuracy


# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression() #LogisticRegression() is a class,reg is object
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
x_train_pred=model.predict(x_train)
train_data=accuracy_score(x_train_pred,y_train)
```

```python
print('accuracy on training data :',train_data)
```

```python
from sklearn.metrics import accuracy_score
x_test_pred=model.predict(x_test)
test_data=accuracy_score(x_test_pred,y_test)
```

```python
print('accuracy on testing data :',test_data)
```

# decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier   
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=classifier.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

# kneighbour classifier

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(x,y)
```

```python
from sklearn.metrics import r2_score
y_pred=classifier.predict(x_test)
r3=r2_score(y_test,y_pred)
print(r3)
```

out of these all model the accuracy of logistic regression is best so we predict our model using logistic regression
